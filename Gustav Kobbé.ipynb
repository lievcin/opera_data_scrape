```python
from bs4 import BeautifulSoup
import requests, re

# name = "Russia/Bolshoi.csv"
# csv = open(name, "w") 
# columnTitleRow = "opera_title\t opera_description\t date\t role\t artist\t url\n"
# csv.write(columnTitleRow)

url = "http://www.gutenberg.org/files/40540/40540-h/40540-h.htm"
r  = requests.get(url)
soup = BeautifulSoup(r.text, "html5lib")

links_with_text = []
for a in soup.find_all('a', href=True): 
    if a.text: 
        links_with_text.append(a['href'])
```

```python
pages_we_dont_want = ['Opera_Before_Gluck', 'Transcribers_Errata_List', 'Contents', 'Page_', 'Transcribers_Errata_List', 'INDEX', 'ILLUSTRATIONS', 'Schools_of_Opera', 'Why_Some_Operas_are_Rarely_Given']
def check_if_wanted(text):
    matched = True
    for page in pages_we_dont_want:
        if text in '#' + page:
            matched = False
            break
    return matched
links_with_text = [link for link in links_with_text if check_if_wanted(link) == True]
links_with_text = [link for link in links_with_text if link.isupper() == True]
links_with_text

# '#ORFEO_ED_EURIDICE'.isupper()

# check_if_wanted('#Transcribers_Errata_List')
```
