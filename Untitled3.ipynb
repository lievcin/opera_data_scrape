```python
from bs4 import BeautifulSoup
import requests
import re
proc_perf = []

base_url = "http://operabase.com/oplist.cgi?id=none&lang=en&adv=f&from=1+1+2016&to=31+12+2020&approx=0&is=&by=&sel=&loc="

for c in cities:
    url = base_url + c + "&near=0&full=Y&sort=V"
#     print(url)
    r  = requests.get(url)
    country_main = BeautifulSoup(r.text, "html5lib")
    tables = country_main.select('table')
    try:
        tables = tables[7]
        rows = tables.find_all('tr')

        for row in rows:
            if row.text != 'DateLocationOpera':
                data = row.find_all('td')[4:]
                theatre = data[0].text
                city = c#data[0].find('span').text
                work = data[1].find('a').text
                date = data[1].find('a')
                date = re.search('\d+', date.get('href'))
                date = date.group(0)
                proc_perf.append([theatre, city, work, date])  
    except:
        print(c)
```

```python
import csv, string
from csv import writer

with open("operabase_from_2016.csv", 'w') as outfile:
    mywriter = writer(outfile)
    mywriter.writerow(['theatre', 'city', 'work', 'date'])
    for perf in proc_perf:
        mywriter.writerow(perf)       
                
```
